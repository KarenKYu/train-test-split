# What could go wrong

### Introduction

In the last lesson, we started with a machine learning algorithm that found the true relationship between customers and temperature beneath our data: $customers = 3*temperature + 10$.  Then we showed that even when the machine learning model finds this true relationship, it cannot perfectly predict noisy data, that is data with random errors.   This is because our machine learning algorithm cannot predict randomness, which makes sense.  Our machine learning algorithm still did as well as possible -- the error resulting from the noise in the data is an *irreducible error*.

Now we'll make things harder for our machine learning algorithm.  Unlike in the last lesson, we won't train our algorithm on a perfectly clean dataset, like we did on the last lesson.

```python
from random import randint, seed
from graph import trace_values, plot
seed(1)
random_temperatures = list(map(lambda num: randint(30,101), range(0, 50)))
perfect_customers = list(map(lambda temp: 3*temp + 10, random_temperatures))
data_trace = trace_values(random_temperatures, perfect_customers, name = 'initial data')
plot([data_trace])
```

In real life, we'll essentially never work with a dataset that perfectly follows a straight line.  As we mentioned before there is a degree of randomness to how many customers visit our restaurant in a given day.  And we need to train the machine learning algorithm on that noisy data.  In this lesson, we'll see the complications that arise when we do so.

### Training on a noisy dataset

In this lesson, we'll work with a scenario that better matches what we would see as data scientists in the real world -- we'll start with noisy data.  Ok, so each day our restaurant tracks the temperature and the number of customers that enter the store.  This is what it sees.

```python
seed(2)
random_errors = list(map(lambda num: randint(-30,30), range(0, 50)))
noisy_customers = list(map(lambda index: perfect_customers[index] + random_errors[index],range(0, 50)))
data_trace = trace_values(random_temperatures, noisy_customers, name = 'initial data')

layout = {'yaxis': {'title': 'customers'}, 'xaxis': {'title': 'temperature'}}
plot([data_trace], layout = layout)
```

Now we do not know the underlying relationship between the temperature and the amount of customers, that's what we use the machine learning algorithm to discover.  So we train our algorithm using this noisy data to try to approach the underlying relationship between temperature and customer amount.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
temperature_inputs = list(map(lambda temperature: [temperature], random_temperatures))
model.fit(temperature_inputs, noisy_customers)
```

```python
model.coef_
```

```python
model.intercept_
```

Now notice that when we train our data on the noisy data, our model came close, but did not discover the true relationship between temperature and customers.  Why not?  Why did it not say that the best fit line is the matches $y = 3x + 10$.  Well it's because when we train the model it is simply trying to draw a line that minimizes the sum of the squared errors through the random data.  And so the model is influenced by the whims and randomness of the data. 

```python
customer_predictions = model.predict(temperature_inputs)
model_trace = trace_values(random_temperatures, customer_predictions, name = 'expected', mode = 'lines')
plot([data_trace, model_trace], layout = layout)
```

Let's see how well the model performs.

```python
model.score(temperature_inputs, noisy_customers)
```

Not bad at all.

### Ship it?

So we show how well this model performs matches the dataset, to the restaurant manager and he is impressed.  We decide to use the the machine learning model to the amount of customers that come into the restaurant in the future.

So here are the new temperatures and new customer amounts that we see enter the restaurant in the following months.

```python
seed(3)

new_random_temperatures = list(map(lambda num: randint(30,101), range(0, 50)))
new_random_errors = list(map(lambda num: randint(-30,30), range(0, 50)))
new_perfect_customers = list(map(lambda temp: 3*temp + 10, new_random_temperatures))

new_noisy_customers = list(map(lambda index: new_perfect_customers[index] + new_random_errors[index],range(0, 50)))
```

And because we do not retrain our model -- we already did that -- we use the same model to predict the future amounts of customers.

```python
data_trace = trace_values(new_random_temperatures, new_noisy_customers, name = 'initial data')

plot([data_trace, model_trace])
```

Ok, it doesn't look bad. Let's see how we did.

```python
new_temperature_inputs = list(map(lambda temperature: [temperature], new_random_temperatures))

model.score(new_temperature_inputs, new_noisy_customers)
```

So our model performed worse on our future data.  And while this is not always the case, when we follow the approach above, this often occurs.  Why does our model perform worse when we put our data into production than when we trained our model?

### This is overfitting

The reason why our model performs worse on the production data that on our training data is because our model adjusted to the randomness on the training set.  So it drew a line to that was influenced by the random fluctuations.  Because these random fluctuations are unlikely to occur in the future, these adjustments hurt the model.  Ideally, the model would have discovered the true relationship of $y = 3x + 10$, but because simply looks at the noisy data and finds a model that minimizes the error, it discovers a different, slightly incorrect model.

This incorrect model overperforms on the training data because it matches the randomness, but it tends to underperform on the production data because matching the randomness of the training data incorrectly influenced the model.  And this phenomenon is reflected in the difference in how the model scores on the training data versus the production data.
