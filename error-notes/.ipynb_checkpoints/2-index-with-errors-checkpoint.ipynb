# What could go wrong

### Introduction

In the last lesson, we saw that given data that is generated by a simple linear regression formula and a set of inputs, our machine learning model can find the parameters of that simple linear regression formula.  

It simply trains against the input data and the ouput data, and it finds the corresponding equation that matches the data.

```python
from random import randint, seed
seed(1)
random_temperatures = list(map(lambda num: randint(30,101), range(0, 50)))
perfect_customers = list(map(lambda temp: 3*temp + 10, random_temperatures))
```

```python
from sklearn.linear_model import LinearRegression
perfect_model = LinearRegression()
temperature_inputs = list(map(lambda temperature: [temperature], random_temperatures))
perfect_model.fit(temperature_inputs, perfect_customers)
perfect_model.coef_
# 3
perfect_model.intercept_
# 10.000000000000028
```

In this lesson, we'll see what happens when we do not observe data that perfectly matches a line.

### A different underlying function

Let's change our underlying machine learning function.

Previously, our underlying formula had been $ customers = 3*temperature + 10$.  Now we'll make it:

$$ customers = 3*temperature + 10 + \epsilon $$ or 
$$ y = 3x + 10 + \epsilon $$

So our formula is the same, but now we're adding some noise to it.  This is how we do this in code.  

We just generate a list of random errors.  Let's generate fifty of them, one for each of our set of $(x,y)$ values, and have them range between positive and negative 30.

```python
random_errors = list(map(lambda num: randint(-30,30), range(0, 50)))
random_errors[0:10]
```

And now we can just add these an error to each of our customer outputs.

```python
customers_with_error = list(map(lambda index: perfect_customers[index] + random_errors[index],range(0, 50)))
```

> So the above line of code, maps through the numbers 0 through 49, and then selects the elements with the 0 index from `perfect_customers` and `random_errors` and adds them together.  Then it does the same with the elements with index 1 from the lists, and so on.

Let's plot our data.

```python
from graph import trace_values, plot
data_trace = trace_values(random_temperatures, customers_with_error, name = 'initial data')

layout = {'yaxis': {'title': 'customers'}, 'xaxis': {'title': 'temperature'}}
plot([data_trace], layout = layout)
```

So as you can see, our data still does follow a trend, after all it is simply following the formula $ y = 3x + 10 + \epsilon $.  However, now our data does not lie perfectly along the line.  Let's call this our *noisy data*.  Another word for error, is noise.

### What does random mean?

Now here, we used a random number generator to add in some randomness to our model.  In the real world, we see randomness as well.  For example, the amount of customers that visit our restaurant in a day will fluctuate based on a customer's mood for our type of food, or if another customer has leftovers in the apartment, or stubbed his toe, or got into an argument and no longer wants to go out.

Now you could argue that one day we could somehow capture and account for each of these scenarios.  However, even if we did, each of these factors is so small, infrequent that it's simply too time consuming to include this in our model.  We call these influences random.  And in the real world, all of our datasets are subject to randomness.

### Using our model on noisy data

Now let's see how well our already trained model can predict random data.  Now we already trained a model on our non-noisy data, and that model discovered the underlying model of y = 3x + 10.  Let's see how well it predicts our noisy data.

```python
customer_predictions = perfect_model.predict(temperature_inputs)
model_trace = trace_values(random_temperatures, customer_predictions, name = 'expected', mode = 'lines')
plot([data_trace, model_trace], layout = layout)
```

```python
perfect_model.score(temperature_inputs, customers_with_error)
```

So it does quite a good job, but obviously worse than when we had a non-noisy dataset.  The question is this: 

*Can we train a machine learning model that does better?*

The answer is, no.  We have a dataset that follows the pattern of $y = 3x + 10 + \epsilon$ and our machine learning algorithm estimates a model of $\hat{y} = 3x + 10$.  It's impossible for the model to predict the random error, because by definition this type of error is random.

In fact another word for this error, is *irreducible error* because this is error that we as data scientists cannot reduce.  We cannot predict a random process.

### Subject to the whims

The last thing to note about this irreducible error, is that we can have a lot of it, or a little of it.  The more random error we have, the smaller our score.  Let's see this.

We'll have our errors range between positive and negative 50 (instead of 30).

```python
larger_random_errors = list(map(lambda num: randint(-50,50), range(0, 50)))
customers_with_larger_error = list(map(lambda index: perfect_customers[index] + larger_random_errors[index],range(0, 50)))
```

```python
perfect_model.score(temperature_inputs, customers_with_larger_error)
```

So we can see that the larger the random error, the less well we can predict the data.

The other thing to remember is that if we want to see the numbers behind these new predictions, we can see them by calling the corresponding methods.
